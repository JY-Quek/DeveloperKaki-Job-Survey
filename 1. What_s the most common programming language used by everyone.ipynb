<h3>What's the most common programmign language used by everyone? </h3>


```python
from tools import data_loader
import pandas as pd
```

<p>Here we import the data_loader module to help us pre_load the data into dataframe and do some cleaning beforehand.</p>

```python
df = data_loader.load_data("Developer Kaki _ Developer Salary Survey 2021 Results.xlsx")
```

```python
df.tech_used  #The column we are interested to investigate
```

<h4>Get the frequency of each tech and sort them from most common to least common

```python
from tools import frequency_counter
```

```python
counter_info, wordcloud = frequency_counter.freq_counter(df)  
# frequency counter will automatically use df.tech_used as reference
```

```python
counter_info.most_common()
```

<h4>From above, we can see that "python" has the highest frequency count. Now, have a quick look at all rows containing the word "python"</h4>

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    display(df[df.tech_used.str.lower().str.strip().str.contains("python")][["tech_used", "salary"]])
```

<h4>Generate a wordcloud for visualisation</h4>

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()  
```
